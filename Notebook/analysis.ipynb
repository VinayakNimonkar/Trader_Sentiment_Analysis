```python
Section 1 — Import Libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
Section 2 — Load Data
sentiment = pd.read_csv("fear_greed_index.csv")
trader = pd.read_csv("historical_data.csv")

sentiment.head()
trader.head()
```

```python
sentiment['date'] = pd.to_datetime(sentiment['date'])
trader['Timestamp IST'] = pd.to_datetime(trader['Timestamp IST'])

trader['date'] = trader['Timestamp IST'].dt.date
sentiment['date'] = sentiment['date'].dt.date
```

```python
Section 4 — Merge Dataset
merged = trader.merge(sentiment[['date','classification']], on='date', how='left')
```

```python
Section 5 — Create Metrics
daily_metrics = merged.groupby(['date','classification']).agg(
    total_pnl=('Closed PnL','sum'),
    avg_pnl=('Closed PnL','mean'),
    trades=('account','count')
).reset_index()

daily_metrics.head()
```

```python
Section 6 — Visualization
sns.barplot(data=daily_metrics, x='classification', y='total_pnl')
plt.title("PnL vs Market Sentiment")
plt.show()
```

```python
Section 7 — Trader Segmentation
trader_perf = merged.groupby('account')['Closed PnL'].sum()

frequent = merged.groupby('account')['account'].count()

segment = pd.DataFrame({
    'pnl': trader_perf,
    'trades': frequent
})

segment.head()
```
